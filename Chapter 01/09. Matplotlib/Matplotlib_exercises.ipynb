### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
# url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

# data = pd.read_csv(url)
data =pd.read_csv('company_sales_data.csv')
data.head()

```

```python
data['total_profit'].plot.line()
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")
plt.show()
```

### Exercise 3: Read all product sales data and show it using a multiline plot.
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

The graph should look like this:


```python
data.head()
```

```python
# Write your code here
fig, ax = plt.subplots(figsize=(10,8))
data['facecream'].plot.line(marker='o',linestyle='-', label='Facecream Sales Data',linewidth=3, markersize=6 )
data['facewash'].plot.line(marker='o',linestyle='-',label='Facewash Sales Data', linewidth=3, markersize=6)
data['toothpaste'].plot.line(marker='o',linestyle='-',label='Toothpaste Sales Data', linewidth=3, markersize=6)
data['bathingsoap'].plot.line(marker='o',linestyle='-',label='Bathingsoap Sales Data', linewidth=3, markersize=6)
data['shampoo'].plot.line(marker='o',linestyle='-',label='Shampoo Sales Data', linewidth=3, markersize=6)
data['moisturizer'].plot.line(marker='o',linestyle='-', label='Moisturizer Sales Data',linewidth=3, markersize=6)
plt.xlabel("Month number",fontsize=15)
plt.ylabel("Sold units in number",fontsize=15)
plt.title("Sale data",fontsize=15)
plt.legend()
plt.rc('legend',fontsize=15)
plt.show()
```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.
Also, add a grid in the plot (the grid line style should be âââ).

The scatter plot should look like this:


```python
# Write your code here
plt.style.use('bmh')
data.plot.scatter(x='month_number',y='toothpaste')
plt.xlabel("Month number",fontsize=12)
plt.ylabel("Number of units sold",fontsize=12)
plt.title("Toothpaste Sale data",fontsize=12)
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
# Write your code here
data.plot(x="month_number", y=["facecream", "facewash"], kind="bar",color=['b', 'r'],label=['Facecream sales','Facewash sales'])
plt.xlabel("Month number",fontsize=12)
plt.ylabel("Sales units in number",fontsize=12)
plt.title("Facewash and Facecream Sale data",fontsize=12)
plt.legend()
plt.rc('legend',fontsize=7)
```

### Exercise 6: Read the sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk.
The bar chart should look like this:


```python
# Write your code here
data.plot(x="month_number", y=["bathingsoap"], kind="bar",legend=None)
plt.xlabel("Month number",fontsize=12)
plt.ylabel("Sales units in number",fontsize=12)
plt.title("Bathingsoap Sale data",fontsize=12)
plt.savefig('Bathingsoap Sale data.png')
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.
The histogram should look like this:


```python
# Write your code here
data.plot(x="month_number", y=["total_profit"], kind="hist")
plt.xlabel("Profit range in dollar",fontsize=12)
plt.ylabel("Actual profit in dollar",fontsize=12)
plt.title("Profit data",fontsize=12)
plt.savefig('Profit data.png')
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
data.head()
```

```python
yearly_unit_sales=(data['facecream'].sum(),data['facewash'].sum(),data['toothpaste'].sum(),data['bathingsoap'].sum(),data['shampoo'].sum(),data['moisturizer'].sum())
sale_labels = 'facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo','moisturizer'
plt.pie(yearly_unit_sales, labels = sale_labels,radius=2 ,autopct = '%1.1f%%')
```


###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
# Write your code here
fig = plt.figure()
fig.add_subplot(2, 1, 1)

data['bathingsoap'].plot.line(marker='o',linestyle='-',label='Bathingsoap Sales Data', linewidth=3, markersize=6)
plt.title('Sales data of a Bathing Soap')

fig.add_subplot(2, 1, 2)
#plt.subplot(2,1,2)
data['facewash'].plot.line(marker='o',linestyle='-',label='Facewash Sales Data', linewidth=3, markersize=6,color ='r')
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title('Sales data of a Facewash')
fig.tight_layout()
```
