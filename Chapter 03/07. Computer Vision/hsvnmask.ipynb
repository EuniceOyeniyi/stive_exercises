```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.figure(figsize=(10,20))
hsv_map = cv2.imread('hsv.png')
# plt.imshow(hsv_map)
hsv_map_rgb = cv2.cvtColor(hsv_map,cv2.COLOR_BGR2RGB )
plt.imshow(hsv_map_rgb)
```

```python
chal_img = cv2.imread('m.jpg')
plt.figure(figsize=(7,25))
rgb_img = cv2.cvtColor(chal_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
hsv_img = cv2.cvtColor(chal_img,cv2.COLOR_BGR2HSV)
```

```python
#Showimg only blues 
lower_bounds = (85,50,10)
upper_bounds = (125,255,255)

mask = cv2.inRange(hsv_img,lower_bounds,upper_bounds)

masked_blues = rgb_img.copy( )
masked_blues[mask==0] = [0,0,0]
plt.figure(figsize=(10,20))
plt.imshow(masked_blues)
```

```python
#showing only greens 
lower_Gbounds = (39,50,10)
upper_Gbounds = (75,255,255)

Gmask = cv2.inRange(hsv_img,lower_Gbounds,upper_Gbounds)

masked_greens = rgb_img.copy( )
masked_greens[Gmask==0] = [0,0,0]
plt.figure(figsize=(10,20))
plt.imshow(masked_greens)
```

```python
#showing only greens and blues
lower_GBbounds = (39,20,0)
upper_GBounds = (125,255,255)

GBmask = cv2.inRange(hsv_img,lower_GBbounds,upper_GBounds)

masked_gb = rgb_img.copy( )
masked_gb[GBmask==0] = [0,0,0]
plt.figure(figsize=(10,20))
plt.imshow(masked_gb)
```

```python
#showing only greens and blues
# lower_GBRbounds = (0,10,0)
# upper_GBRounds = (125,200,250)
lower_GBRbounds = (0,10,0)
upper_GBRounds = (125,210,250)
GBRmask = cv2.inRange(hsv_img,lower_GBRbounds,upper_GBRounds)

masked_gbR = rgb_img.copy( )
masked_gbR[GBRmask==0] = [0,0,0]
plt.figure(figsize=(10,25))
plt.imshow(masked_gbR)
```

```python
final_RGB = cv2.addWeighted(rgb_img,0.6,masked_gbR,0.8,0)
plt.imshow(final_RGB)
```
