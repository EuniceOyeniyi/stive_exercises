```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
path = 'C:\\Users\\halle\Documents\\Datasets\\07.Computer Vision Course Resources\\images\\dog.jpg'

dog = cv2.imread(path)
image = cv2.cvtColor(dog,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,25))
plt.imshow(image)
```

```python
gray_dog = cv2.cvtColor(dog, cv2.COLOR_BGR2GRAY)
retval, dst = cv2.threshold(gray_dog, 200, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(10,25))
# plt.imshow(gray_dog,cmap='gray')
plt.imshow(dst, cmap='gray')
```
